```python
dataset = pd.read_csv('/content/drive/MyDrive/CSV files/UMTMVS.csv', index_col='DATE', parse_dates=True)
dataset.index
```

```python
dataset.loc['2000-01-01':'2015-01-01']
```

```python
dataset.loc['1992-01-01':'2000-01-01':12]
```

```python
#time resampling
dataset.resample(rule='AS').mean().head()
```

```python
#Rolling time series
dataset.rolling(window=10).mean().head(20) # head to see first 20 values 
```

```python
dataset.rolling(window=10).mean().tail(20) # tail to see last 20 values 
```

```python
#maximum value from a window of 30 days
dataset.rolling(window=30).max()[30:].head(20) # head is just to check top 20 values
```

```python
#Plotting timeseries data using pandas

dataset.plot(color = '#A61F69')
plt.grid()
```

```python
dataset.plot(title='Value of Manufacturers Shipments', ylabel='Value',color='#BE0000',figsize=(12,10),grid='axis')
```

```python
dataset.resample(rule='AS').mean().plot(title='Value of Manufacturers Shipments', ylabel='Value',color='#F2921D',figsize=(12,10),grid='axis',linewidth='5')

```

```python
dataset.resample(rule='AS').mean().plot.bar(title='Value of Manufacturers Shipments', ylabel='Value',color='#9DF1DF',figsize=(12,10),grid='axis')

```

```python
dataset['UMTMVS'].resample(rule='MS').mean().plot(color='#206A5D',figsize=(15,8), label='Resample MS')
plt.autoscale(tight=True)
dataset.rolling(window=30).mean()['UMTMVS'].plot(ylabel='Value of Mean of Starting of Month',title='Average of Manufacturers Shipments',label='Rolling window=30',grid='axis',linewidth='5',color='#FFB830')
plt.legend()
```

```python
#plotting for specific dates as per your choice.
dataset['UMTMVS'].resample(rule='AS').max().plot(xlim=["1999-01-01","2014-01-01"],ylim=[280000,540000], figsize=(12,7),grid='axis',color='#FF7878',linewidth='5')
```
